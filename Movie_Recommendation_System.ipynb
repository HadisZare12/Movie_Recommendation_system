```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#loading the data from the csv file
movies_data = pd.read_csv('/content/movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
#selecting the relevant features for system
selected_features = ['genres','keywords','tagline','cast','director']
```

```python
#replacing the null val with null string
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
#combinig the selected_features
combined_features = movies_data['genres']+ ' '+movies_data['keywords']+ ' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
#converting text to vector
vectorizer = TfidfVectorizer()
```

```python
feature_vector = vectorizer.fit_transform(combined_features)
```

```python
#similarity
similarity = cosine_similarity(feature_vector)
```

```python
#getting the movie name from user
movie_name = input('Enter your fav movie : ')
```

```python
#creating the list of movies
list_of_titles = movies_data['title'].tolist()
```

```python
#find the close match
find_close_match=difflib.get_close_matches(movie_name,list_of_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
```

```python
#find the index of movie
index_of_movie =movies_data[ movies_data.title == close_match]['index'].values[0]
print(index_of_movie)
```

```python
similarity_score= list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
#sorting the movies based on their score
sorted_similar_movie = sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movie)
```

```python
#print the name of similar movie
print('movie suggested for you : ')

i = 1
for movie in sorted_similar_movie:
  index = movie[0]
  title_from_index = movies_data[movies_data.index == index]['title'].values[0]
  if(i<30):
    print(i,'.',title_from_index)
    i+=1
```

```python
#summerize the system
movie_name = input(' Enter your favourite movie name : ')

list_of_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_titles)

close_match = find_close_match[0]

index_of_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
